```python
bm = SMPL(model_name='smpl', gender='neutral')
```

```python
fitter = SMPLfit(bm, num_betas=10)
```

```python
# seed:
torch.manual_seed(3)
beta = torch.randn(1, 10)
pose = torch.randn(1, 72) * 0.2
transl = torch.randn(1, 3) * 0.1
res = bm(pose, beta, transl)


```

```python
f = 1.5
fit_res = fitter.fit_with_known_pose(pose, res['vertices']*f, res['joints']*f, scale_fit=True, beta_regularizer=0)
res2 = bm(pose, fit_res['shape_betas'])
#fit_res['scale_corr'] = 1
s = fit_res['scale_corr']
res2['vertices'] =  res2['vertices'] * s + fit_res['trans']
res2['joints'] =  res2['joints'] * s +  fit_res['trans']

error = (res2['vertices']/f - res['vertices']).norm(dim=-1).mean(dim=1)
print(error.numpy(), fit_res['scale_corr'],  s)
```

```python
fit_smplx = converter_smpl2smplx.convert(fit_res['pose_rotvecs'], fit_res['shape_betas'], fit_res['trans'], kid_factor=torch.ones_like(fit_res['pose_rotvecs'][:, :1]))
fit2_smplx = converter_smpl2smplx.convert(fit_res['pose_rotvecs'], fit_res['shape_betas'], fit_res['trans'], kid_factor=torch.ones_like(fit_res['pose_rotvecs'][:, :1]), known_output_pose_rotvecs=fit_smplx['pose_rotvecs'])
print(fit2_smplx['shape_betas'] - fit_smplx['shape_betas'])
```
